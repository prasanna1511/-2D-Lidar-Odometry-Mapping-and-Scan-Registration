```python
import numpy as np
import struct
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the path to your specific .bin file
bin_file = "data/BINARY/00000.bin"  # Adjust the file name accordingly

# Open the .bin file in binary mode
with open(bin_file, "rb") as f:
    # Read all data from the file
    data = f.read()

    # Assuming each point is represented by 3 floats (x, y, z) in binary format
    num_points = len(data) // (3 * 4)  # 3 floats (4 bytes each)

    # Unpack binary data assuming little-endian format and float data type
    points = struct.unpack("<{}f".format(num_points * 3), data)

    # Reshape the data into a NumPy array of shape (num_points, 3)
    points = np.array(points).reshape(-1, 3)

# Visualize the point cloud using Matplotlib
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:, 0], points[:, 1], points[:, 2], c='b', marker='.', s=1)  # Adjust marker size as needed
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Point Cloud Visualization')

plt.show()
```
